# **Aim:-** Template matching techniques for object detection

# **Theory**

Template matching is a technique in digital image processing used to find small parts of an image that match a template image.

It works by sliding the template image over the input image and calculating a metric that indicates the degree of similarity between the template and the region under the template at each location.

The `cv2.matchTemplate` function in OpenCV takes the input image, template image, and method of comparison as arguments. It returns a grayscale image where each pixel denotes the match value at that location.


image.png


There are different comparison methods available, such as:

* cv2.TM_CCOEFF_NORMED: Normalized Cross Correlation.
* cv2.TM_SQDIFF: Sum of Squared Differences.
* cv2.TM_CCORR: Cross Correlation.
And others.

Once you have the result, you can use cv2.minMaxLoc to find the location with the minimum or maximum value, depending on the comparison method. This location corresponds to the top-left corner of the best match. You can then draw a rectangle around the matched region using cv2.rectangle.

Different methods work better in different scenarios, and the choice depends on the specific application.

# **Code:**

Import required libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/content/drive/MyDrive/CVR Lab/boy_image.jpg')
```

```python
#Convert to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
img_crop = cv2.imread('/content/drive/MyDrive/CVR Lab/boy_crop_template.jpg')
```

```python
#Convert to RGB
img_crop = cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_crop)
```

```python
img_crop.shape
```

```python
template_height, template_width, channels = img_crop.shape
```

```python
template_height
```

```python
template_width
```

```python
channels
```

```python
result = cv2.matchTemplate(img, img_crop, cv2.TM_CCOEFF_NORMED)
```

```python
plt.imshow(result)
```

Find the location with the highest correlation

```python
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
```

```python
print(min_val)
print(max_val)
print(min_loc)
print(max_loc)
```

Get the top-left corner of the best match

```python
top_left = max_loc
top_left
```

Draw a rectangle around the matched area

```python
bottom_right = (top_left[0] + template_width, top_left[1] + template_height)
bottom_right
```

```python
img_detected_template = cv2.rectangle(img, top_left, bottom_right, (255, 255, 0), 20)
```

```python
plt.imshow(img_detected_template)
```
